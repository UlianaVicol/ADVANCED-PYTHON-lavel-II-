```python
# Object <class dict>
class User:
    # common logic (python specific)
    def __init__(self, nickname, password):
        self.nickname = nickname
        self.password = password
        self.online   = False
        self.friends  = [] # list of User objects
        self.posts    = [] # list of Post objects
        self.messages  = []
        self.directMessage = []
        self.status_message = "unseen"

    # User specific LogIn (autentific)  
    def __str__(self):
        return f"User <{self.nickname}>"
    
    def __repr__(self):
        return self.__str__()
    
    # User specific LogIn (autentific)
    def auth(self , nickname , password):
        if nickname == self.nickname and password == self.password:
            self.online = True 
            # lose coupling 
            return True
        else:
            return False
    def addFriend(self ,friend):
        # direct reference
        # HW2: check if the friend type is User class
        if  type(friend) != User:
            #self.friends.append(friend)
            return
        friend_exists = False
        # HW1: check if the frend Doesn't exist yet!!! search a value in the list 
        #      for + if -> nickkname, break
        for existing_friend in self.friends:
            if existing_friend.nickname == friend.nickname:
                friend_exists = True
                break
        if not friend_exists:
            self.friends.append(friend)

    def post(self, title, body):
        #HW4: chech if both title and body are of type string
        if type(title) == str and type(body) == str:
            new_post = Post(title,body)
            self.posts.append(new_post)
            new_post.author = self
        return new_post

    def process_message_success(self, body , toUser):
        message = Message(self.nickname ,body ,toUser )
        mess_box = str(message).replace(f" - to > :< {toUser.nickname} >" , "")
        sent_mes = str(message).replace(f" < {self.nickname} >" , "")
        self.messages.append(f" |#|Sent Message|#|   - <{sent_mes}>")
        toUser.directMessage.append(f"|#|Recieved Message|#| - < {mess_box} >")
        return message

    def read_message(self, message_box):
        status = self.directMessage[message_box]
        changestatus = status.replace("unseen" , "message seen")
        self.directMessage.pop(message_box)
        self.directMessage.insert(message_box, changestatus)
        return changestatus    
```

```python
class Post:
    def __init__(self, title , body):
        self.title = title
        self.body  = body
    def __str__(self):
        return f"{self.title}"
    def __repr__(self):
        return self.__str__()
    
class Message:
    def __init__(self, fromUser, body , toUser):
        self.body = body
        self.status = "unseen"
        self.fromUser = fromUser
        self.toUser = toUser
    def __str__(self):
        return f" ####### Message From #######  |#|< {self.fromUser} >|#|  < {self.body} > |   <<|| {self.status} ||>> - to > :< {self.toUser.nickname} >"
    
    
    def __repr__(self):
        return self.__str__()
```
